# numpy random

```python
import numpy as np
```

### create between [0,1)

```python
print(np.random.rand())               # float
print(np.random.rand(5))              # numpy.ndarray
print(np.random.rand(3,3))            # numpy.ndarray
```

### create sample   mean: 0   std: 1

```python
print(np.random.randn())               # float
print(np.random.randn(5))              # numpy.ndarray
print(np.random.randn(3,3))            # numpy.ndarray
```

### create between [0,1)

```python
print(np.random.random())             # float
print(np.random.random(5))            # numpy.ndarray
print(np.random.random([3,3]))        # numpy.ndarray
```

### create normal sample

```python
print(np.random.normal(65, 3))           # float
print(np.random.normal(65, 3, 5))        # numpy.ndarray
print(np.random.normal(65, 3, [3,3]))    # numpy.ndarray
```

### create random integer between [low, high)

```python
print(np.random.randint(0,10))       # int
print(np.random.randint(0,10,5))       # numpy.ndarray
print(np.random.randint(0,10,[3,3]))       # numpy.ndarray
```
